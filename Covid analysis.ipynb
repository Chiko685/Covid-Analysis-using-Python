```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
#this module is used to create more complex and customized visualizations by combining multiple plots into a single figure.
```

```python
def country_visualisation(df, country_name):
    country_data = df[df['Country/Region'] == country_name]
    fig = make_subplots(rows=1, cols=4, subplot_titles=('Date', 'TotalDeaths', 'TotalRecovered', 'ActiveCases'))
    
    fig.add_trace(go.Scatter(x=country_data['Date'], y=country_data['Confirmed'], mode='lines+markers', name='Total Confirm'), row=1, col=1)
    fig.add_trace(go.Scatter(x=country_data['Date'], y=country_data['Deaths'], mode='lines+markers', name='Total Deaths'), row=1, col=2)
    fig.add_trace(go.Scatter(x=country_data['Date'], y=country_data['Recovered'], mode='lines+markers', name='Total Recovered'), row=1, col=3)
    fig.add_trace(go.Scatter(x=country_data['Date'], y=country_data['Active'], mode='lines+markers', name='Active Cases'), row=1, col=4)
    
    fig.update_layout(title_text=f'COVID-19 Trends in {country_name}', template='plotly_dark', height=600, width=700)
    fig.show()
```

```python
country_visualisation(full_group, 'Brazil')
```
